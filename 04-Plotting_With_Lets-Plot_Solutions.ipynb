```python
import pandas as pd
import numpy as np
from palmerpenguins import load_penguins
from lets_plot import *

LetsPlot.setup_html()
penguins = load_penguins()
```

## Exercise 1

(a) What is wrong with this code - can you fix it to make all points blue?

```python
(
    ggplot(data=penguins)
    + geom_point(aes(x='body_mass_g', y='flipper_length_mm', color="blue"))
    + labs(title="Penguins Body Mass vs Flipper Length (Fixed Color)")
)
```

The color parameter specified inside `aes()` expects a variable (in our case, a column name) in order to assign different colors of the points based on the value of this variable. Since there is no such column called "blue" it throws an error. What we instead want is to set all points in the plot to blue, so we have to define the color outside the `aes()` call:

```python
(
    ggplot(data=penguins)
    + geom_point(aes(x='body_mass_g', y='flipper_length_mm'), color="blue")
    + labs(title="Penguins Body Mass vs Flipper Length (Fixed Color)")
)
```

(b) We want to create a plot that shows the `flipper_length_mm` (y-axis) based on the `body_mass_g` (x-axis). The size of the points should depend on `flipper_length_mm`. Fix this code.

```python
(
    ggplot(data=penguins)
    + geom_point(aes(x='body_mass_g', y='flipper_length_mm'), size="flipper_length_mm")
)
```

Here, it is just the other way round: the size of the points to plot should depend on `flipper_length_mm` and therefore has to be inside the `aes()` call.

```python
(
    ggplot(data=penguins)
    + geom_point(aes(x='body_mass_g', y='flipper_length_mm', size="flipper_length_mm"))
)
```
